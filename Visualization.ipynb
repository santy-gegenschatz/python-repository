```python
# Importaciones
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats
from datetime import datetime
```

## Ejercicio 1

```python
# Cargamos el file csv mediante pandas
df = pd.read_csv('hr-dataset.csv')
```

```python
# También es interesante obtener la mediana y la moda de variables como el salario
mediana = np.median(df['Salary'])
print(f'La mediana de los salarios es {mediana}')
moda = stats.mode(df['Salary'])
moda = moda[0][0]
print(f'La moda de los salarios es {moda}')
```

## Ejercicio 2

```python
# Cambiamos el estilo de matplotlib
mpl.style.use('bmh')
```

```python
# Realizamos un histograma con matplotlib
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 12))

ax[0].hist(df['Salary'], bins = 20)
ax[1].scatter(df['Salary'], df['Absences'])

ax[0].set_title('Histograma de los salarios')
ax[0].set_xlabel('USD')
ax[0].set_ylabel('Frecuencia')
xinterval = np.arange(45000,250000,10000)
ax[0].set_xticks(xinterval)

ax[1].set_title('Relación entre salarios y ausencias')
ax[1].set_xlabel('Salarios')
ax[1].set_ylabel('Ausencias')
```

```python
print('La mayoría de los salarios se encuentran entre los 50 y los 100K.')
print('En particular hay un pico alrededor de los 55 y 65K.')
```

## Ejercicio 3

```python
# Hacer un gráfico de violín de los salarios, usando seaborn
mpl.style.use('default')
ax = sns.catplot(data = df, kind = 'violin', x = 'Sex', y = 'Salary', hue ='MarriedID', split = True)
ax.set(title = 'Gráfico de violin - Salarios según el sexo', xlabel = 'Sexo', ylabel = 'Salario')
```

## Ejercicio 4

```python
# Graficamos la serie de tiempo correspondiente a Date of Hire.
date_of_hire = df['DateofHire']
group = df.groupby(date_of_hire).count()
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (12,7))
ax.plot(group.index, group['EmpID'])
ax.set_title('Fecha de Contratación')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Cantidad de Contratados')
ax.xaxis.set_major_locator(plt.MaxNLocator(9))
```
